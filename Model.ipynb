```python
def show_metrics(y_test, y_pred):
  tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
  print(f'True positive: {tp}\nTrue negative: {tn}\nFalse positive: {fp}\nFalse negative: {fn}\n')
  print("==================================")
  print(f'recall: {recall_score(y_test, y_pred)}')
  print(f'precision: {precision_score(y_test, y_pred)}')
  print(f'f1: {f1_score(y_test, y_pred)}')
  print("==================================\n")
```

```python
def model_fit_predict(model):
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  show_metrics(y_test, y_pred)

def select_features(model):
  selector = SelectFromModel(model)
  selector.fit(X_train, y_train)
  status = selector.get_support()
  all_features = X_train.columns
  selected_features = X_train.columns[status].tolist()
  print(f"All features number: {len(all_features)}, Selected features number: {len(selected_features)}, Selected_features:")
  print(selected_features)
  return selected_features
```

# Feature importance

```python
def count_features():
  counted = []
  dict_of_features = {}
  list_of_features = feat_logreg + feat_ridge + feat_linsvc + feat_ranfor + feat_dectre
  for feature in list_of_features:
    if feature not in counted:
      number = list_of_features.count(feature)
      dict_of_features[feature] = number
      counted.append(feature)
  return sorted(dict_of_features.items(), key=lambda item: item[1])
```

```python
features = count_features()
features
```

```python
chosen_features = [item[0] for item in features if item[1]>=3]
chosen_features
```

# Comparison of important features for all and paying students

```python
df_students_paying = df_students[df_students['Paying']==1]
df_students_paying
```

```python
df_students_free = df_students[df_students['Paying']==0]
df_students_free
```

```python
def means_comparison():
  records = []
  for col in chosen_features:
    free_mean = df_students_free[col].mean()
    pay_mean = df_students_paying[col].mean()
    diff = pay_mean - free_mean
    columns = {'Feature': col, 'Free_accounts': free_mean, 'Paying_students': pay_mean, 'Difference': diff}
    records.append(columns)
  df = pd.DataFrame(records)
  return df
```

```python
df = means_comparison()
df
```

```python
def medians_comparison():
  records = []
  for col in chosen_features:
    free_median = df_students_free[col].median()
    pay_median = df_students_paying[col].median()
    diff = pay_median - free_median
    columns = {'Feature': col, 'Free_students': free_median, 'Paying_students': pay_median, 'Difference': diff}
    records.append(columns)
  df = pd.DataFrame(records)
  return df
```

```python
df = medians_comparison()
df
```

```python
def max_comparison():
  records = []
  for col in chosen_features:
    free_max = df_students_free[col].max()
    pay_max = df_students_paying[col].max()
    diff = pay_max - free_max
    columns = {'Feature': col, 'Free_students': free_max, 'Paying_students': pay_max, 'Difference': diff}
    records.append(columns)
  df = pd.DataFrame(records)
  return df
```

```python
df = max_comparison()
df
```
